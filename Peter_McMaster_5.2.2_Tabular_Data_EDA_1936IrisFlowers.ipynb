```python
# Load the Data

from sklearn.datasets import load_iris

iris = load_iris()
iris
```

```python
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create pandas DataFrame

iris_df = pd.DataFrame(data = iris.data,
                       columns = iris.feature_names)
```

```python
# Add column with Target Variable

target = iris.target
iris_df["Target"] = target
```

```python
# Add Column with Target Names

target_names = iris.target_names
names = []
for i in range(len(iris.target)):
    names.append(target_names[iris.target[i]])
iris_df["Species"] = names
```

```python
# Show iris_df

iris_df
```

```python
iris_df.describe()
```

```python
iris_df.Species.value_counts()
```

```python
# 2-D scatter plot

iris_df.plot(kind='scatter', x='sepal length (cm)', y='sepal width (cm)');
plt.show()
```

```python
# Scatter Plot with Hue for Species
# sepal length VS sepal width

sns.set_style("whitegrid");
sns.FacetGrid(iris_df, hue="Species", height=4, aspect=2) \
    .map(plt.scatter, "sepal length (cm)", "sepal width (cm)") \
    .add_legend()
plt.show()
```

```python
# Scatter Plot with Hue for Species
# petal length VS petal width

sns.set_style("whitegrid");
sns.FacetGrid(iris_df, hue="Species", height=4, aspect=2) \
    .map(plt.scatter, "petal length (cm)", "petal width (cm)") \
    .add_legend()
plt.show()
```
